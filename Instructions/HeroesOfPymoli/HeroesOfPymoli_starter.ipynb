```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
#Because some SNs occur more than once in this data, we want to count only the unique values
total_players = len(purchase_data['SN'].unique())
total_players_s = pd.Series({"Total Players": total_players}).to_frame()

total_players_s
```

```python
#total unique items
unique_items = len(purchase_data['Item ID'].unique())
#average price
Avg_price = purchase_data["Price"].mean()
Avg_price = "${:,.2f}".format(Avg_price)
#unique purchases
total_trans = len(purchase_data["Purchase ID"])
#total revenue of sales
total_price = purchase_data["Price"].sum()
total_price = "${:,.2f}".format(total_price)

#put in a dataframe
summary_stat_table = pd.DataFrame({"Number of Unique Items": unique_items,
                   "Average Price": [Avg_price] ,
                   "Number of Purchases":[total_trans],
                    "Total Revenue": [total_price]})
summary_stat_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#group by SN and Gender and get the count for each SN
gender_data = purchase_data[["SN","Gender","Price"]].groupby(["SN","Gender"])
gender_dedup = pd.DataFrame(gender_data["Price"].count())

#reset the index to pull Gender back to a column, then group by Gender
gender_count = gender_dedup.reset_index(drop = False)
gender_count = gender_count.groupby(["Gender"])

#get the count of unique screen names by gender, rename price value, calculate percent, sort descending
gender_count_summary = pd.DataFrame(gender_count["Price"].count())
gender_count_summary = gender_count_summary.rename(columns = {"Price":"Total Count"})
gender_count_summary["Percentage of Players"] = (gender_count_summary["Total Count"]/total_players*100).map("{:,.2f}".format)
gender_count_summary.sort_values("Total Count", ascending = False)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Counting all the purchases by Gender, including duplicate SNs
gender_dups_data = purchase_data[["Gender","Price"]].groupby(["Gender"])
gender_dups_count = pd.DataFrame(gender_dups_data["Price"].count())
gender_dups_count = gender_dups_count.reset_index(drop = False).rename(columns = {"Price":"Purchase Count"})

#sum Price over SN, group by Gender, and then sum by Gender, take Gender out of index, and rename price value 
gender_dedup_sum = pd.DataFrame(gender_data["Price"].sum())
gender_sum = gender_dedup_sum.reset_index(drop = False).groupby(["Gender"])
gender_sum_summary = pd.DataFrame(gender_sum["Price"].sum())
gender_sum_summary = gender_sum_summary.reset_index(drop = False).rename(columns = {"Price":"Total Purchase Value"})
gender_sum_summary["Total Purchase Value"] = gender_sum_summary["Total Purchase Value"].map("${:,.2f}".format)

#use the summed by SN, then sum by gender, take Gender out of index, and rename price value
gender_avg_summary = pd.DataFrame(gender_sum["Price"].mean())
gender_avg_summary = gender_avg_summary.reset_index(drop = False).rename(columns = {"Price":"Avg Total Purchase per Person"})
gender_avg_summary["Avg Total Purchase per Person"] = gender_avg_summary["Avg Total Purchase per Person"].map("${:,.2f}".format)

#mean Price over SN, group by Gender, and then mean by Gender, take Gender out of index, and rename price value 
gender_dedup_mean = pd.DataFrame(gender_data["Price"].mean())
gender_avg = gender_dedup_mean.reset_index(drop = False).groupby(["Gender"])
gender_avg_avg = pd.DataFrame(gender_avg["Price"].mean())
gender_avg_avg = gender_avg_avg.reset_index(drop = False).rename(columns = {"Price":"Average Purchase Price"})
gender_avg_avg["Average Purchase Price"] = gender_avg_avg["Average Purchase Price"].map("${:,.2f}".format)

#Combine each summary statistic df into one big df
gender_merge = pd.merge(gender_dups_count, gender_avg_avg, on = "Gender")
gender_merge = pd.merge(gender_merge, gender_sum_summary, on = "Gender")
gender_merge = pd.merge(gender_merge, gender_avg_summary, on = "Gender").set_index("Gender")
gender_merge
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create bins and labels to categorize age groups
bins = [0,9,14,19,24,29,34,39,45]
group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#adding the Age Group category to the data and grouping by SN and Age Group
age_data = purchase_data
age_data["Age Group"] = pd.cut(age_data["Age"], bins, labels=group_names)
age_data_grouped = age_data[["SN","Age Group","Price"]].groupby(["SN","Age Group"])

#by counting with SNs grouped together, we only count each SN once
age_dedup = pd.DataFrame(age_data_grouped["Price"].count())
age_count = age_dedup.reset_index(drop = False).rename(columns = {"Price":"Total Count"})

#group the data (with SN dedupped), and then counthow many unique SNs are in each Age Group
age_count = age_count.groupby(["Age Group"])
age_count_summary = pd.DataFrame(age_count["Total Count"].count())
age_count_summary["Percentage of Players"] = (age_count_summary["Total Count"]/total_players*100).map("{:,.2f}".format)
age_count_summary.sort_values("Total Count", ascending = False)

age_count_summary
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group the data by Item ID and Item Name
item_data = purchase_data[["Item ID","Item Name","Price"]]
item_data_grouped = item_data.groupby(["Item ID","Item Name"])

#Count of purchase for each item
item_count = pd.DataFrame(item_data_grouped["Price"].count())
item_count = item_count.reset_index(drop = False).rename(columns = {"Price":"Purchase Count"})

#total cost of all purchases for each item
item_sum = pd.DataFrame(item_data_grouped["Price"].sum())
item_sum = item_sum.reset_index(drop = False).rename(columns = {"Price":"Total Purchase Value"})
item_sum = item_sum[["Item ID","Total Purchase Value"]]

#price per item
item_avg = pd.DataFrame(item_data_grouped["Price"].mean())
item_avg = item_avg.reset_index(drop = False).rename(columns = {"Price":"Item Price"})
item_avg = item_avg[["Item ID","Item Price"]]

#merge information from summary df into one big df
item_merge = pd.merge(item_count, item_avg, on = "Item ID")
item_merge = pd.merge(item_merge, item_sum, on = "Item ID")

#sort item df by Purchase Count
item_merge_sort = item_merge.sort_values("Purchase Count", ascending = False).set_index("Item ID")

#formatting
item_merge_sort["Total Purchase Value"] = item_merge_sort["Total Purchase Value"].map("${:,.2f}".format)
item_merge_sort["Item Price"] = item_merge_sort["Item Price"].map("${:,.2f}".format)
item_merge_sort.head()
```
